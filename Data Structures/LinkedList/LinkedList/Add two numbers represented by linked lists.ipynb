```python
class Node(object):
    def __init__(self, val):
        self.data = val
        self.next = None
        
class Linkedlist():
    def __init__(self):
        self.head = None
    
    def traverse(self):
        temp = self.head
        while temp :
            print(temp.data, end = " ")
            temp = temp.next
        print()
        
def num_to_linkedlist(num):
    list1 = Linkedlist()
    for i in num:
        k = Node(int(i))
        if list1.head == None:
            list1.head = k
        else:
            k.next = list1.head
            list1.head = k

    return list1

def add_linkedlist(list1, list2):
    temp1, temp2 = list1.head, list2.head
    if not temp1:
        return temp2
    
    if not temp2:
        return temp1

    list3 = Linkedlist()
    carry = 0
    while temp1 and temp2 :
        summ = temp1.data + temp2.data + carry
        if summ >= 10:
            carry = int(str(summ)[0])
            num = int(str(summ)[1])
            if list3.head == None:
                list3.head = Node(num)
            else:
                k = Node(num)
                k.next = list3.head
                list3.head = k
        else:
            num = int(str(summ)[0])
            if list3.head == None:
                list3.head = Node(num)
            else:
                k = Node(num)
                k.next = list3.head
                list3.head = k
    
        temp1 = temp1.next
        temp2 = temp2.next
        
        
    while temp1:
        k = Node(temp1.data + carry)
        k.next = list3.head
        list3.head = k
        temp1 = temp1.next
        carry = 0
        
    while temp2:
        k = Node(temp2.data + carry)
        k.next = list3.head
        list3.head = k
        temp2 = temp2.next
        carry = 0
    
    if carry > 0 :
        new = Node(carry)
        new.next = list3.head
        list3.head = new
        
    return list3
            

l1 = num_to_linkedlist(input())        
l2 = num_to_linkedlist(input())
sum_ = add_linkedlist(l1, l2)
sum_.traverse()
```
