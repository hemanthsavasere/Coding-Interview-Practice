```python
class Node(object):
    def __init__(self,val):
        self.data = val
        self.next = None
    
class LinkedList(object):
    def __init__(self):
        self.head = None

    def traverse(self):
        temp = self.head
        while temp:
            print(temp.data, end = " ")
            temp = temp.next
        print()
        
def merge(list1, list2):
    l3 = LinkedList()
    
    if list1.head == None:
        return list2
    
    elif list2.head == None:
        return list1
    
    else:
        temp1, temp2 = list1.head, list2.head
        temp3 = l3.head
        
        while temp1 and temp2:
            
            if temp1.data <= temp2.data:
                k = Node(temp1.data)
                temp1 = temp1.next
                
            else:
                k = Node(temp2.data)
                temp2 = temp2.next
        
        
            if not temp3:
                l3.head = k
                temp3 = k
              
            else:
                temp3.next = k
                temp3 = temp3.next
            
        while temp1:
            k = Node(temp1.data)
            temp3.next = k
            temp1 = temp1.next
            temp3 = temp3.next
           
        while temp2:
            k = Node(temp2.data)
            temp3.next = k
            temp2 = temp2.next
            temp3 = temp3.next
        
        return l3
            
            
        
l1 = LinkedList()
l2 = LinkedList()

n1 = Node(10)
n2 = Node(20)
n3 = Node(30)
n4 = Node(40)

n1.next = n2
n2.next = n3
n3.next = n4

l1.head = n1

k1 = Node(15)
k2 = Node(21)
k3 = Node(34)
k4 = Node(43)

l2.head = k1

k1.next = k2
k2.next = k3
k3.next = k4

l1.traverse()
l2.traverse()

print(merge(l1, l2).traverse())

```
